## <font color='#e7c299'>1. Deep Learning</font> <a id="eje0">




La unidad fundamental de procesamiento en las redes neuronales es **el Perceptrón**, el cual puede ser utilizado como un algoritmo de inteligencia artificial en determinados casos. Este componente se basa en la manera en que las neuronas biológicas operan y fue concebido por Frank Rosenblatt en 1957.


<div align="center"><img src="https://i.imgur.com/zgZNocs.png" ></div>

Cada neurona biológica consta de varias partes, entre ellas, el cuerpo celular que contiene la mayoría de los componentes complejos de la célula y el núcleo, además de extensiones ramificadas denominadas dendritas, así como una extensión larga llamada axón. En el extremo del axón, se bifurca en telodendrias y, en sus puntas, se encuentran estructuras diminutas llamadas terminales sinápticos o sinapsis, las cuales se conectan con otras neuronas en sus dendritas o cuerpos celulares. Las neuronas biológicas generan breves impulsos eléctricos conocidos como potenciales de acción o señales que recorren el axón, produciendo la liberación de neurotransmisores en las sinapsis. Si una neurona recibe suficientes neurotransmisores en un lapso de tiempo breve, dispara sus propios impulsos eléctricos. Pese a que las neuronas individuales parecen actuar de manera sencilla, pueden conectarse con miles de otras neuronas para crear redes complejas capaces de llevar a cabo tareas complicadas.

### **Componentes de una red neuronal (Perceptron)**

####  <font color="#2C6E49">***Capas***</font>

*  Capas de entrada: es la información de entrada, esta debe ser un vector

*  Capas ocultas: son las conexiones internas en la red, donde a cada neurona se le asignara un peso para calcular la salida

*  Capas de salida: es la informacion de salida calcula a partir de la red.

<div align="center"><img src="https://www.atriainnovation.com/wp-content/uploads/2019/10/Redes_neuronales_esquema.png" ></div>

El numero de capas a usar depende de la complejidad del problema a resolver, por ejemplo, si se quiere clasificar imagenes de perros y gatos, se necesitan muchas capas para poder identificar las caracteristicas de cada una de las clases.

### <font color="#2C6E49">***Neuronas***</font>

son las unidades de procesamiento fundamental, cada neurona tiene una entrada, una salida y una función de activación asociada.

Para calcular la salida de una neurona se realiza la siguiente operación:

$$ \color{orange} { \hat{y} = \sum_{i=1}^{n} w_i x_i + b}$$

donde $w_i$ es el peso de la entrada $x_i$, $b$ es el sesgo y $n$ es el numero de entradas.

Esto tambien se puede ver como una multiplicación de matrices:

$$ \color{orange} {\hat{y} = \mathbf{W} \cdot \mathbf{X} + \mathbf{b}}$$


<div align="center"><img src="https://i.imgur.com/a4b7UBT.png" width="400" ></div>

donde $\mathbf{W}$ es la matriz de pesos, $\mathbf{X}$ es el vector de entradas y $\mathbf{b}$ es el vector de sesgos.

### <font color="#2C6E49">***Funciónes de activación***</font>

Es la función que se aplica a la salida de la neurona para obtener la salida final de la red, esta función debe ser no lineal para poder resolver problemas complejos.

<font color='#A792CE'>¿Por qué no lineal?</font>

Si la función de activación es lineal, la red neuronal se comporta como una regresión lineal, es decir, la salida de la red es una combinación lineal de las entradas, por lo que no se puede resolver problemas complejos.

```python
# forward-pass of a 3-layer neural network:
import numpy as np

f = lambda x: 1/(1+np.exp(-x)) # sigmoid function
X = np.random.randn(1, 6) # random input (6x1)

W1 = np.random.randn(6, 4) # random weights layer 1 (6x4)
b1 = np.random.randn(1, 4) # random bias layer 1 (1x4)

W2 = np.random.randn(4, 3) # random weights layer 2 (4x3)
b2 = np.random.randn(1, 3) # random bias layer 2 (1x3)

W3 = np.random.randn(3, 1) # random weights layer 3 (3x1)
b3 = np.random.randn(1, 1) # random bias layer 3 (1x1)

# forward pass
h1 = f(X.dot(W1) + b1)
h2 = f(h1.dot(W2) + b2)
y = f(h2.dot(W3) + b3)

print(y)
```

### **Funciones de pérdida**

Valor verdadero $ = \mathbf{y}$

Valor predicho $ =\mathbf{\hat{y}}$

*  **Error cuadrático medio (MSE)**: calcula la cantidad de error que hay entre dos conjuntos de datos.

$$ \color{orange}  {MSE = \frac{1}{N}\sum_{i=1}^{N}(y_{i}-\hat{y_{i}})^2}$$


*  **Binary Crossentropy** : calcula la pérdida de entropía cruzada entre etiquetas verdaderas y etiquetas predichas. Se usa para aplicaciones de clasificación binaria (0 o 1). 

$$ \color{orange} { BC = -\frac{1}{N}\sum_{i=1}^{N}[y_{i}\cdot log(\hat{y_{i}})+(1-y_{i})\cdot log(1-\hat{y_{i}})]}$$


### **Backpropagation**

El backpropagation es un algoritmo que permite calcular los gradientes de la función de pérdida con respecto a los pesos de la red neuronal, esto permite actualizar los pesos de la red para minimizar la función de pérdida.

en la red queremos **minimizar** la función de pérdida:

$$ \color{orange}  {W* = argmin_{W} \mathcal{L}(W)}$$

para esto se calcula el gradiente de la función de pérdida con respecto a los pesos de la red, esto se hace de manera recursiva:

se calcula el gradiente de la función de pérdida con respecto a los pesos de la última capa, luego se calcula el gradiente de la función de pérdida con respecto a los pesos de la penúltima capa y así sucesivamente hasta llegar a la primera capa.

el gradiente se calcula de la siguiente manera:

$$ \color{orange}  {\nabla_{W}  = \frac{\partial \mathcal{L}(W)}{\partial W}}$$

donde $W$ es la matriz de pesos de la red.




### **Optimizadores**

Los optimizadores son los encargados de actualizar los pesos de la red, para esto se utiliza el gradiente de la función de pérdida con respecto a los pesos de la red.

Para este taller utilizaremos el optimizador **SGD** (Stochastic Gradient Descent), es el optimizador más simple, actualiza los pesos de la red de la siguiente manera:

$$ \color{orange}  { W_{t+1} = W_{t} - \alpha \nabla_{W}}$$

donde $W_{t}$ son los pesos de la red en el paso $t$, $\alpha$ es el learning rate y $\nabla_{W} es el gradiente de la función de pérdida con respecto a los pesos de la red.


### **Métricas de evaluación**

Las métricas de evaluación son las encargadas de medir el desempeño de la red neuronal, para este taller utilizaremos las siguientes métricas:

*  **Accuracy**: es la cantidad de predicciones correctas que hace la red.

$$ \color{orange}  {Accuracy = \frac{TP+TN}{TP+TN+FP+FN}}$$

*  **Precision**: es la cantidad de predicciones positivas que son correctas.

$$ \color{orange}  {Precision = \frac{TP}{TP+FP}}$$

*  **Recall**: es la cantidad de predicciones positivas que se hicieron correctamente.

$$ \color{orange}  {Recall = \frac{TP}{TP+FN}}$$

TP = True Positive (predicción positiva correcta)

TN = True Negative (predicción negativa correcta)

FP = False Positive (predicción positiva incorrecta)

FN = False Negative (predicción negativa incorrecta)

##  <font color='#e7c299'> 3. Ejercicio básico de denoising </font>  <a name="eje2"> 


```python
#@title **Código:** Agregar ruido a los datos
#@markdown Aplicamos un ruido con una distribución normal a la base de datos, con un factor de 0.5.
#@markdown Imprimimos los primeros 10 valores de la base de datos una vez aplicado el ruido.
noise_factor = 0.3
x_train_noisy = x_train + noise_factor * np.random.normal(loc=0.0, scale=1.0, size=x_train.shape) 
x_test_noisy = x_test + noise_factor * np.random.normal(loc=0.0, scale=1.0, size=x_test.shape) 

x_train_noisy = np.clip(x_train_noisy, 0., 1.)
x_test_noisy = np.clip(x_test_noisy, 0., 1.)

n = 10
plt.figure(figsize=(20, 2))
for i in range(1, n + 1):
    ax = plt.subplot(1, n, i)
    plt.imshow(x_test_noisy[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
plt.show()
```

```python
#@title **Código:** Definir modelo con arquitectura autoencoder
#@markdown <img src="https://i.ibb.co/XjHpsyD/0-uq2-Zip-B9-Tq-I9-G-k.png" width="900">
input_img = keras.Input(shape=(28,28,1))

input = layers.Flatten()(input_img)
encoded = layers.Dense(128, activation='relu')(input)
encoded = layers.Dense(64, activation='relu')(encoded)

encoded = layers.Dense(32, activation='relu')(encoded)

decoded = layers.Dense(64, activation='relu')(encoded)
decoded = layers.Dense(128, activation='relu')(decoded)
decoded = layers.Dense(784, activation='sigmoid')(decoded)

out = layers.Reshape((28, 28, 1), input_shape=(784,))(decoded)

autoencoder = keras.Model(input_img, out)

autoencoder.summary()
```

```python
#@title **Código:** Entrenar el modelo
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')

history = autoencoder.fit(x_train_noisy, x_train,
                epochs=50,
                batch_size=256,
                shuffle=True,
                validation_data=(x_test_noisy, x_test))
```

```python
#@title **Código:** Curvas de convergencia
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(['Función de pérdida: entrenamiento','función de pérdida: validación'])
plt.xlabel('Épocas')
plt.xlabel('Binary crossentropy')
plt.title('Entrenamiento autoencoder')
plt.show()
```

```python
#@title **Código:** Resultados obtenidos de los datos reconstruidos.
#@markdown Imprimimos los primeros 10 valores de los datos orignales, los datos con ruido, y los datos reconstruidos.

decoded_imgs = autoencoder.predict(x_test_noisy)

n = 10
plt.figure(figsize=(20, 2))
for i in range(1, n + 1):
    ax = plt.subplot(1, n, i)
    plt.imshow(x_test[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
plt.show()

plt.figure(figsize=(20, 2))
for i in range(1, n + 1):
    ax = plt.subplot(1, n, i)
    plt.imshow(x_test_noisy[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
plt.show()

plt.figure(figsize=(20, 2))
for i in range(1, n + 1):
    ax = plt.subplot(1, n, i)
    plt.imshow(decoded_imgs[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
plt.show()
```

```python
#@title **Código:** Función para las métricas.
def mse(vref, vcmp):
    if np.iscomplexobj(vref) or np.iscomplexobj(vcmp):
        dtype = np.complex128
    else:
        dtype = np.float64
    r = np.asarray(vref, dtype=dtype).ravel()
    c = np.asarray(vcmp, dtype=dtype).ravel()
    return np.mean(np.abs(r - c)**2)

def psnr(vref, vcmp, rng=None):
    if rng is None:
        rng = np.abs(vref.max() - vref.min())
    dv = (rng + 0.0)**2
    with np.errstate(divide='ignore'):
        rt = dv / mse(vref, vcmp)
    return 10.0 * np.log10(rt)

def snr(vref, vcmp):
    dv = np.var(vref)
    with np.errstate(divide='ignore'):
        rt = dv / mse(vref, vcmp)
    return 10.0 * np.log10(rt)

print('Métricas de evaluación de la calidad de la reconstrucción')
print('MSE =', np.round(mse(x_test[7,:,:], decoded_imgs[7,:,:]),5))
print('PSNR =', np.round(psnr(x_test[7,:,:], decoded_imgs[7,:,:]),5))
print('SNR =', np.round(snr(x_test[7,:,:], decoded_imgs[7,:,:]),5))

plt.imshow(decoded_imgs[7,:,:,0]);
```

## <font color='#e7c299'> 4. Modelos Convolucionales </font> <a name="eje3"> 

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.
x_train = np.expand_dims(x_train, -1)
x_test = np.expand_dims(x_test, -1)
```

```python
#@title **Código:** Definir Modelo con arquitectura de Red Neuronal Convolucional (CNN)
#@markdown <img src="https://research.binus.ac.id/airdc/wp-content/uploads/sites/18/2018/03/ConvNet-1-640x288.png" width="900">



input_img = keras.Input(shape=(28, 28, 1))
nclasses = 10

x = layers.Conv2D(16, (3, 3), activation='relu', padding='same')(input_img)
x = layers.MaxPooling2D((2, 2), padding='same')(x)

x = layers.Conv2D(16, (3, 3), activation='relu', padding='same')(x)
x = layers.MaxPooling2D((2, 2), padding='same')(x)

x = layers.Flatten()(x)
classes = layers.Dense(nclasses, activation='softmax')(x)

CNN = keras.Model(input_img, classes)
CNN.summary()
```

**Convoluciones**

Estas consisten en tomar «grupos de pixeles cercanos» de la imagen de entrada e ir operando matemáticamente (producto escalar) contra una pequeña matriz que se llama kernel.
