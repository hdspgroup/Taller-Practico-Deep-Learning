#**Vectorized code vs For loop**

![Vectorized vs Scalar](https://nccastaff.bournemouth.ac.uk/jmacey/Lectures/SIMD/images/simd1.png)

```python
w = np.random.rand(1000000)
x = np.random.rand(1000000)
tic = time.time()
c = np.dot(w,x)

toc = time.time()
print('Vectorized version: '+ str(1000*(toc-tic))+' ms')
c = 0

tic = time.time()
for i in range(1000000):
    c += w[i]*x[i]

toc = time.time()
print("For loop: ",str(1000*(toc-tic))+' ms')
```
